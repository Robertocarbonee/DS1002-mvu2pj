## Pie Chart (3 points)

```python
# Create a pie chart from raw values

slices = [56, 66, 24, 124, 47]
labels = ["US","UK","MX","CA","JP"]
```

```python
import matplotlib.pyplot as plt

labels = ["US","UK","MX","CA","JP"]
sections = [56, 66, 24, 124, 47]
colors = ['r', 'g', 'yellow']

plt.pie(sections, labels=labels, colors=colors,
        startangle=90,
        autopct = '%1.2f%%')

plt.axis('equal')
plt.title('Pie Chart ')
plt.show()
```

## Scatterplot with Trendline (3 points)

```python
# Create a scatterplot with a trendline from two lists of values

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

num_points = 50
x = 5 + np.arange(num_points) + np.random.randn(num_points)
y = 10 + np.arange(num_points) + 5 * np.random.randn(num_points)

# If you want them combined into a dataframe
df = pd.DataFrame({'x': x, 'y': y})
```

```python
plt.scatter(df['x'], df['y'], color= 'black', label='Data Points')

z = np.polyfit(df['x'], df['y'], 1)
p = np.poly1d(z)
plt.plot(df['x'], p(df['x']), color='red', label='Trendline')
plt.title('Scatterplot with Trendline')
plt.show()
```
